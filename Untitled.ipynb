```python
from transformers import AutoTokenizer, AutoModel
import numpy as np

def get_embeddings_all_models(sentence1, sentence2, models):
    embedding_results = {}

    for model_name in models:
        print(f"\nUsing model: {model_name}")
        tokenizer = AutoTokenizer.from_pretrained(model_name)
        model = AutoModel.from_pretrained(model_name)

        # Compute embeddings
        q1_emb = get_embeddings(sentence1, model_name, tokenizer, model)
        q2_emb = get_embeddings(sentence2, model_name, tokenizer, model)

        # Store embeddings
        embedding_results[model_name] = (q1_emb, q2_emb)

        # Print preview
        print("  - q1 embedding (first 5 dims):", q1_emb[:5])
        print("  - q2 embedding (first 5 dims):", q2_emb[:5])

    return embedding_results
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Labels and data
models = ['Similarity Ensemble', 'Concatenated Embeddings', 'CCA Ensemble']
precision = [0.80, 0.61, 0.60]   # Weighted avg precision
recall =    [0.86, 0.97, 0.99]   # Weighted avg recall
accuracy =  [0.79, 0.60, 0.59]  # Overall accuracy

x = np.arange(len(models))
width = 0.25

# Plotting
fig, ax = plt.subplots(figsize=(10, 6))
bars1 = ax.bar(x - width, precision, width, label='Precision (Weighted Avg)', color='skyblue')
bars2 = ax.bar(x, recall, width, label='Recall (Weighted Avg)', color='lightgreen')
bars3 = ax.bar(x + width, accuracy, width, label='Accuracy', color='salmon')

# Formatting
ax.set_xlabel('Model')
ax.set_ylabel('Score')
ax.set_title('Precision, Recall, and Accuracy Across Models')
ax.set_xticks(x)
ax.set_xticklabels(models, rotation=10)
ax.set_ylim(0, 1.25)
ax.legend()
ax.grid(True, axis='y', linestyle='--', alpha=0.6)

# Annotations
for bar in bars1 + bars2 + bars3:
    height = bar.get_height()
    ax.annotate(f'{height:.2f}', xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 3), textcoords="offset points", ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
